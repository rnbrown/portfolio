```python
import pandas as pd
spray=pd.read_csv('../../input/clean/spray_cleaned.csv').drop(['Unnamed: 0'],axis=1)
train=pd.read_csv('../../input/clean/train_cleaned.csv').drop(['Unnamed: 0'],axis=1)
weather=pd.read_csv('weather_agged.csv').drop(['Unnamed: 0'],axis=1)
test=pd.read_csv('../../input/clean/test_cleaned.csv').drop(['Unnamed: 0', 'Id'],axis=1)
```

```python
weather['Date']=pd.to_datetime(weather['Date'])
test['Date']=pd.to_datetime(test['Date'])
train['Date']=pd.to_datetime(train['Date'])
spray['Date']=pd.to_datetime(spray['Date'])
```

```python
def spray_factor(row):
    Date = row.Date
    Lat = row.Lat
    Long = row.Long   
    rel=spray[(spray['Date']<Date) & ((Date-spray['Date'])<pd.Timedelta('14 days'))]
    s=0
    for _ in rel.iterrows():
        latdiff=Lat-_[1]['Lat']
        longdiff=Long-_[1]['Long']
        dis=.5**(latdiff**2+longdiff**2)
        t=(Date-_[1]['Date']).days
        s+=(.5)**(t/1.5)/dis
    return s
```

```python
train['spray_factor'] = train.apply(spray_factor, axis=1)
train_df=train.merge(weather, on='Date')
test['spray_factor'] = test.apply(spray_factor, axis=1)
test_df=test.merge(weather, on='Date')
```
