```python
def eda(dataframe):
    print;print "missing values \n", dataframe.isnull().sum()
    print;print "dataframe index \n", dataframe.index
    print;print "dataframe types \n", dataframe.dtypes
    print;print "dataframe shape \n", dataframe.shape
    print;print "dataframe describe \n", dataframe.describe()
    print "Unique count"
    for item in dataframe:
        print;print item
        print dataframe[item].nunique()
        
eda(train)
```

```python
### This is real
train_agrigated = train.copy()

# train_agrigated['NumMosquSum'] = 

# df['Total'] = df.groupby(['Fullname', 'Zip'])['Amount'].transform('sum')
```

```python
#
#
# Sum the number of Mosquitos found per Day per Trap per Species, also sum if WNV was found more than once
#
#      WNV / NumMosqSum should be a meaningfull metric for rate of ocuurance
#
train_agrigated[['NumMosqSum', 'WnvSum']] = train.groupby(['Date', 'Species', 'Address', 'Trap'])\
                                                 ['NumMosquitos', 'WnvPresent'].transform('sum')
```

```python
# Looking at the duplicate values now, you can see we have origional rows and the aggriages
#
train_agrigated[train_agrigated.duplicated(['Date', 'Species', 'Address', 'Trap'], keep=False)]\
                                .sort_values(['Date', 'Species', 'Address', 'Trap', 'NumMosquitos', 'WnvPresent'])
```

```python
# Drop the NumMos col and WnvPres col, as they are meaningless in this df
#
train_agrigated.drop(['NumMosquitos', 'WnvPresent'], axis=1, inplace=True)
```

```python
#after groupby the df still has the origonal rows and the group by, dropping dupes fiexs that
#
train_agrigated.drop_duplicates(inplace=True)
```

```python
# Total number of Mosquitos found per week, origonal data
train_agrigated[['NumMosqSum']].groupby([train_agrigated["Date"].dt.year,
                               train_agrigated["Date"].dt.week]).sum().plot(kind="bar",figsize=(15, 5))

# Total batches of Wnv found per week, origional data
train_agrigated[['WnvSum']].groupby([train_agrigated["Date"].dt.year,
                               train_agrigated["Date"].dt.week]).sum().plot(kind="bar",color='orange',figsize=(15, 5))

# plt.show()
```

```python
# Total number of Mosquitos found per week, aggrigated data
train['NumMosquitos'].groupby([train["Date"].dt.year,
                               train["Date"].dt.week]).sum().plot(kind="bar",figsize=(15, 5))

# Total batches of Wnv found per week, aggrigated data
train[['WnvPresent']].groupby([train["Date"].dt.year,
                               train["Date"].dt.week]).sum().plot(kind="bar",color='orange',figsize=(15, 5))
```
